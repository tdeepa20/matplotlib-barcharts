# Matplotlib - Practice Tasks
This notebook contains practice examples for creating and working with matplotlib bar chart.

## Bar chart
      -A bar chart in Matplotlib is a type of chart used to display categorical data using rectangular bars. 
      -The height (or length) of each bar represents the value of each category

**Dataset**:Population by City

## Types
     -Vertical bar chart
     -Horizontal bar chart

### Vertical bar chart
      -Vertical bars are the default type of bar chart in Matplotlib, created using the function plt.bar()

### Horizontal bar chart
      -Horizontal bar charts are created using plt.barh() in Matplotlib.
      -They display data across the y-axis using horizontal bars, with length representing the value.

### Different methods of datatypes
      1.List
      2.Series
      3.Dataframe
      4.Numpy
      5.Function
      6.Dictionary

#### 1.Bar chart using list

```python
import matplotlib.pyplot as plt
```

```python
cities = ['city A','city B','city C','city D']
population = [2.6,3.0,3.1,2.7] #in millions

plt.bar(cities,population, color="green",width=0.5)
plt.xlabel("cities")
plt.ylabel("population")
plt.title("Population by City")
plt.savefig("barchart.png")
plt.show()
```

#### 2.Bar chart using Series

```python
import pandas as pd
```

```python
import numpy as np
```

```python
cities = pd.Series(['city A','city B','city C','city D'])
population =pd.Series ([2.6,3.0,3.1,2.7]) #in millions

plt.bar(cities,population, color="red",width=0.5)
plt.xlabel("cities")
plt.ylabel("population")
plt.title("Population by City")
plt.yticks(np.arange(0,10,2))
plt.show()
```

```python
population =pd.Series (data=[2.6,3.0,3.1,2.7], 
                        index=['city A','city B','city C','city D']) #in millions

population.plot(kind="bar",color="orange")
plt.xlabel("cities")
plt.ylabel("population")
plt.title("Population by City")
plt.yticks(np.arange(0,10,2))
plt.xticks(rotation=0)
plt.savefig("barchart1.png")
plt.show()
```

#### 3.Bar chart using Dataframe

```python
cities = ['city A','city B','city C','city D']
population = [2.6,3.0,3.1,2.7]

df= pd.DataFrame({"Nameofcities":cities,"populationpercity":population})
df.set_index("Nameofcities", inplace=True)
df.plot(kind="bar",color="skyblue")
plt.xlabel("cities")
plt.ylabel("population")
plt.title("Population by City")
plt.yticks(np.arange(0,10,2))
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### 4.Bar chart using Numpy

##### Horizontal bar

```python
cities =('city A','city B','city C','city D')
y_axis=np.arange(len(cities))
population =(2.6,3.0,3.1,2.7) #in milllions

plt.barh(y_axis,population, color="red",align="center",height=0.4,alpha=0.5)
plt.xlabel("population")
plt.ylabel("cities")
plt.title("Population by City")
plt.yticks(y_axis,cities)
plt.xticks(np.arange(0,5,1))
plt.show()
```

#### Bar chart using Function

```python
def fplot(cities,population):
    plt.bar(cities,population,color="purple",width=0.4,alpha=0.5)
    plt.xlabel("cities")
    plt.ylabel("population")
    plt.title("Population by City")
    plt.show()

cities = ['city A','city B','city C','city D']
population = [2.6,3.0,3.1,2.7]
fplot(cities,population)
```

#### Bar chart using Dictionary

```python
popul_data = {"city A": 2.6, "city B": 3.0,"city C": 3.1,"city D": 2.7}

cities = list(popul_data.keys())
population = list(popul_data.values())

plt.bar(cities,population,color='magenta',width=0.4,alpha=0.5)
plt.xlabel("cities")
plt.ylabel("population")
plt.title("Population by City")
plt.show()
```

#### Multiple Bars in One Chart

```python
cities = ['city A','city B','city C','city D']
populationof2015= [2.6,4.0,6.1,8.7]
populationof2020 = [1.4,1.6, 3.0, 2.8]
populationof2025 = [0.4,0.6,1.5, 1.8]


plt.bar(cities,populationof2015, label="2015",color='green',width=0.4 )
plt.bar(cities,populationof2020, label="2020",color='darkblue',width=0.4 )
plt.bar(cities,populationof2025, label="2025",color='orange',width=0.4 )
plt.xlabel("cities")
plt.ylabel("population")
plt.title("Population by City")
plt.legend()
plt.show()
```

#### Styling and Customization

```python
plt.bar(x,y) 
     #Draws a bar chart using x and y data.
plt.barh(y,x) 
     #Draws a horizontal bar chart using x and y data.
plt.xlabel() 
     #Adds a label to the X-axis
plt.ylabel()
     # Adds a label to the Y-axis
plt.xticks()
     # Sets custom tick values or labels on the X-axis
plt.yticks()
     # Sets custom tick values on the Y-axis
plt.legend()
     # Displays the legend for labeled bars
plt.title()
     # Sets the title at the top of the chart
plt.tight_layout()
     # Adjust spacing so nothing gets cut off 
plt.show()
     # Displays the final chart on screen
plt.grid(True)
      # Adds grid to chart for better readability
plt.savefig("barchart.png")
     # Saves the chart as an image file
```

```python
color="red"
      # Sets the bar color to red
width=0.4
      # Controls thickness of vertical bars
edgecolor='black'
      # Adds border color around bars
align='center'
      # Aligns bars relative to ticks
alpha=0.7
      # Sets transparency of bars (0 = transparent, 1 = opaque)
```

### Full Bar Chart with All Styling Options Applied

```python
popul_data = {"city A": 2.6, "city B": 3.0,"city C": 3.1,"city D": 2.7}

cities = list(popul_data.keys())
population = list(popul_data.values())

plt.bar(cities,population,color='orange',width=0.4,edgecolor='black',align="center",alpha=0.5)
plt.xlabel("cities")
plt.ylabel("population")
plt.title("Population by City")
plt.xticks(rotation=45)
plt.yticks(np.arange(0,5,1))
plt.tight_layout()
plt.grid(True)
plt.savefig("barchart3.png")
plt.show()
```
